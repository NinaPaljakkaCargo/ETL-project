```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load
life_data = "lifeData.csv"

# Read Data File and store into Pandas DataFrames
life_df = pd.read_csv(life_data)
life_df.head()
```

```python
life_clean = life_df.drop('Total expenditure',axis =1)
life_clean = life_clean.drop('percentage expenditure',axis =1)
life_clean = life_clean.drop('Income composition of resources',axis =1)
```

```python
life_clean = life_clean.drop('Alcohol',axis =1)
life_clean = life_clean.drop('under-five deaths ',axis=1)
life_clean = life_clean.drop(' thinness  1-19 years',axis =1)
life_clean = life_clean.drop(' thinness 5-9 years',axis =1)

life_clean.head()
```

```python
life_clean.to_csv('life_exp_clean.csv')
```

```python
#mortality and life expectancy by country, including country status by developed(ing), GDP, schooling (2015)
#only year 2015
country_status_clean = life_clean[life_clean.Year == 2015]
#drop infant deaths, virus data, BMI
country_status_clean = country_status_clean.drop('infant deaths',axis=1)
country_status_clean = country_status_clean.drop('Hepatitis B',axis=1)
country_status_clean = country_status_clean.drop('Measles ',axis=1)
country_status_clean = country_status_clean.drop(' BMI ',axis=1)
country_status_clean = country_status_clean.drop('Polio',axis=1)
country_status_clean = country_status_clean.drop('Diphtheria ',axis=1)
country_status_clean = country_status_clean.drop(' HIV/AIDS',axis=1)

country_status_clean.head()
```

```python
country_status_clean.to_csv('country_status_clean.csv')
```

```python
#data for developing countries in 2015
developing_clean = country_status_clean[country_status_clean.Status == 'Developing']

developing_clean.head(12)
```

```python
developing_clean.to_csv('developing_clean.csv')
```

```python
#developed countries in 2015
developed_clean = country_status_clean[country_status_clean.Status == 'Developed']

developed_clean.head(12)
```

```python
developed_clean.to_csv('developed_clean.csv')
```
