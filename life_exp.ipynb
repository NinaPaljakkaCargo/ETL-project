```python
# Dependencies and Setup
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
```

```python
# File to Load
life_data = "lifeData.csv"

# Read Data File and store into Pandas DataFrames
life_df = pd.read_csv(life_data)
life_df.head()
```

```python
life_clean = life_df.drop('Total expenditure',axis =1)
life_clean = life_clean.drop('percentage expenditure',axis =1)
life_clean = life_clean.drop('Income composition of resources',axis =1)
```

```python
life_clean = life_clean.drop('Alcohol',axis =1)
life_clean = life_clean.drop('under-five deaths ',axis=1)
life_clean = life_clean.drop(' thinness  1-19 years',axis =1)
life_clean = life_clean.drop(' thinness 5-9 years',axis =1)

life_clean.head()
```

```python
life_clean.to_csv('life_exp_clean.csv')
```

```python
#data for developing countries, all years (2000-2015)

#df for all years
country_stats = life_clean.drop('infant deaths',axis=1)
country_stats = country_stats.drop('Hepatitis B',axis=1)
country_stats = country_stats.drop('Measles ',axis=1)
country_stats = country_stats.drop(' BMI ',axis=1)
country_stats = country_stats.drop('Polio',axis=1)
country_stats = country_stats.drop('Diphtheria ',axis=1)
country_stats = country_stats.drop(' HIV/AIDS',axis=1)

country_stats.head()

developING_00_15 = country_stats[country_stats.Status == 'Developing']

developING_00_15.head()
```

```python
developING_00_15.to_csv('developING_00_15.csv')
```

```python
#data for developed countries, all years (2000-2015)

#df for all years
country_stats_2 = life_clean.drop('infant deaths',axis=1)
country_stats_2 = country_stats_2.drop('Hepatitis B',axis=1)
country_stats_2 = country_stats_2.drop('Measles ',axis=1)
country_stats_2 = country_stats_2.drop(' BMI ',axis=1)
country_stats_2 = country_stats_2.drop('Polio',axis=1)
country_stats_2 = country_stats_2.drop('Diphtheria ',axis=1)
country_stats_2 = country_stats_2.drop(' HIV/AIDS',axis=1)

developED_00_15 = country_stats_2[country_stats_2.Status == 'Developed']

developED_00_15.head()
```

```python
developED_00_15.to_csv('developED_00_15.csv')
```

```python
#mortality and life expectancy by country, including country status by developed(ing), GDP, schooling (2015)
#only year 2015
country_status_15 = life_clean[life_clean.Year == 2015]
#drop infant deaths, virus data, BMI
country_status_15 = country_status_15.drop('infant deaths',axis=1)
country_status_15 = country_status_15.drop('Hepatitis B',axis=1)
country_status_15 = country_status_15.drop('Measles ',axis=1)
country_status_15 = country_status_15.drop(' BMI ',axis=1)
country_status_15 = country_status_15.drop('Polio',axis=1)
country_status_15 = country_status_15.drop('Diphtheria ',axis=1)
country_status_15 = country_status_15.drop(' HIV/AIDS',axis=1)

country_status_15.head()
```

```python
country_status_15.to_csv('country_status_15.csv')
```

```python
#data for developing countries in 2015
developing_15 = country_status_15[country_status_15.Status == 'Developing']

developing_15.head()
```

```python
developing_15.to_csv('developing_15.csv')
```

```python
#developed countries in 2015
developed_15 = country_status_15[country_status_15.Status == 'Developed']

developed_15.head(12)
```

```python
developed_15.to_csv('developed_15.csv')
```
